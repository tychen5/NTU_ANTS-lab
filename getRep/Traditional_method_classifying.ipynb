```python
% run FeatureTrace.ipynb
def getFeatureProfiles(filePaths):
    featureObj = FeatureTrace
    featureProfiles = dict()
    for f in filePaths:
        if not os.path.isdir(f):
            fName = os.path.basename(f)
#             featureProfile = featureObj(f).getOrignialTrace_withoutTS_inSet()
            featureProfile = featureObj(f).getFeaturedTrace_noTS_inSet()
            featureProfiles[fName] = featureProfile
    return featureProfiles
```

```python
train_labels = []
test_labels = []

for familyName in sorted(train_data.keys()):
    fileCount = 0
    tracePaths = train_data[familyName]
    
    for tracePath in tracePaths:
        if os.path.isfile(tracePath):
            train_labels.append(familyName.split('/')[-2])
            fileCount += 1
            
for familyName in sorted(test_data.keys()):
    fileCount = 0
    tracePaths = test_data[familyName]
    
    for tracePath in tracePaths:
        if os.path.isfile(tracePath):
            test_labels.append(familyName.split('/')[-2])
            fileCount += 1
```

```python
X_Train = df.as_matrix() # trainin data
Y_Train = np.array(train_labels) # training label
X_Test = testDF.as_matrix() # testing data
Y_Test = np.array(test_labels) # testing label
```

```python
### create Sample Name Index

sampleIndex = list()
for traceName in df.index:
    sampleName = traceName.split('_')[0]
    sampleIndex.append(sampleName)

### create Sample Label List Index
labelList = sorted(list(set(Y_Train)))
labelIDX = {l:idx for idx, l in enumerate(labelList , start=1)}

# Training set labels
new_Y_Train = []
for ytra in Y_Train:
    new_Y_Train.append(labelIDX[ytra])
new_Y_Train = np.array(new_Y_Train)

# Testing set labels
new_Y_Test = []
for ytra in Y_Test:
    new_Y_Test.append(labelIDX[ytra])
new_Y_Test = np.array(new_Y_Test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
def getSamplePNR(sampleIndex, y_true, y_pred):
    lastSample = ""
    lastTrueLabel = ""
    isMatch = False
    matchResult = dict()
    predList = []
    samplePredDict = dict()
    for idx, pred in enumerate(y_pred):
        currentSample = sampleIndex[idx]
        trueLabel = y_true[idx]
        if currentSample != lastSample:
            if idx != 0:
                samplePredDict[lastSample] = (lastTrueLabel, predList)
                predList = []
                isMatch = False

            if pred == trueLabel:
                matchResult[currentSample] = 'Match'
                isMatch = True
            else:
                matchResult[currentSample] = 'Mismatch'

            lastSample = currentSample
            lastTrueLabel = trueLabel
        else:
            if not isMatch:
                if pred == trueLabel:
                    matchResult[currentSample] = 'Match'
                    isMatch = True

        predList.append(pred)

    samplePredDict[lastSample] = (lastTrueLabel, predList)

    my_true = []
    my_pred = []
    for sha, labels in samplePredDict.items():
        trueLabel, predList = labels
    #     print("==> true:", trueLabel)
    #     print(predList)
        if trueLabel in predList:
            my_pred.append(trueLabel)
        else:
            my_pred.append(predList[0])
        my_true.append(trueLabel)

    print("accuracy rate:",accuracy_score(my_true, my_pred))
    print("--Micro-Average--")
    print('micro precesion:',precision_score(my_true, my_pred, average='micro'))
    print('micro recall:',recall_score(my_true, my_pred, average='micro'))
    print('micro f1:',f1_score(my_true, my_pred, average='micro'))

    print("--Macro-Average--")
    print('macro precesion:',precision_score(my_true, my_pred, average='macro'))
    print('macro recall:',recall_score(my_true, my_pred, average='macro'))
    print('macro f1:',f1_score(my_true, my_pred, average='macro'))   
    
    print("--Weighted-Average--")
    print('weighted precesion:',precision_score(my_true, my_pred, average='weighted'))
    print('weighted recall:',recall_score(my_true, my_pred, average='weighted'))
    print('weighted f1:',f1_score(my_true, my_pred, average='weighted'))   
```

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel='rbf', C=100)
svm_classifier.fit(X_Train, new_Y_Train)
svm_pred = svm_classifier.predict(X_Test)

y_true = new_Y_Test
y_pred = svm_pred
print("--SVM--")
getSamplePNR(sampleIndex, y_true, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB


gaussianNB_classifier=GaussianNB()
gaussianNB_classifier.fit(X_Train, Y_Train)
gsNB_pred = gaussianNB_classifier.predict(X_Test)

y_true = Y_Test
y_pred = gsNB_pred
print("--Gaussian NB--")
getSamplePNR(sampleIndex, y_true, y_pred)
```

```python
from sklearn.naive_bayes import BernoulliNB

bernoulliNB_classifier=BernoulliNB()
bernoulliNB_classifier.fit(X_Train, Y_Train)
bnNB_pred = bernoulliNB_classifier.predict(X_Test)

y_true = Y_Test
y_pred = bnNB_pred
print("--Bernoulli NB--")
getSamplePNR(sampleIndex, y_true, y_pred)
```
